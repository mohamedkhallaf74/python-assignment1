```python
import pandas as pd
dataset=pd.read_csv("Wuzzuf_Jobs.csv")
dataset.describe()
```

```python
dataset.info();
```

```python
#Having too many slices, when it’s generally best to limit the segments in a pie to 5
x=dataset['Company'].value_counts()
import matplotlib.pyplot as plt
plt.pie(x)
plt.show() 
```

```python
dataset.shape
```

```python
 #Return DataFrame with duplicate rows removed
dataset.drop_duplicates(inplace=True)
dataset.dropna(inplace=True)
 dataset.info();   
```

```python
dataset.shape
```

```python
Jobs_companies = dataset['Company'].value_counts()[:5]
print(Jobs_companies)
```

```python

plt.subplots(figsize=(30,10))
companies = ['Confidential','Mishkat Nour', 'Expand Cart', 'EGIC', 'Aqarmap.com']
plt.pie(Jobs_companies, labels=companies,autopct ='%1.2f%%' )
plt.title("Jobs_companies")
plt.legend();
```

```python
x = dataset['Title'].value_counts()[:10]
print(x)
```

```python
plt.subplots(figsize=(10,10))
plt.barh(x.index,x,height = 0.5, color='red',  linewidth=0.2);
plt.title("MOST POPULAR JOBS")
plt.ylabel("JOBS")
plt.xlabel("DEMAND");
```

```python
q= dataset['Location'].value_counts()[:10]
print(q)
```

```python
plt.subplots(figsize=(15,7))
plt.bar(q.index,q,width = 0.4, color='green');
plt.title("MOST POPULAR LOCATIONS")
plt.ylabel("RATE")
plt.xlabel("CITIES");
```

```python
z = dataset['Skills'].value_counts()[:5]
print(z)
```

```python
plt.subplots(figsize=(30,10))
groups = ['Group_1','Group_2', 'Group_3', 'Group_4', 'Group_5']
plt.pie(z, labels=groups,autopct ='%1.2f%%' )
plt.title("Skills");
plt.legend();
```

```python
# split column into multiple columnsby delimiter 
Seperated_Skills= dataset['Skills'].str.split(',', expand=True)
Seperated_Skills.head(5)
```

```python
#rename columns : 
Seperated_Skills.columns = ['S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13', 'S14', 'S15', 'S16', 'S17']
Seperated_Skills.head(5)
```

```python
s1 = Seperated_Skills['S1'].value_counts()[:10]
print(s1)
plt.subplots(figsize=(10,10))
plt.barh(s1.index,s1,height = 0.5, color='magenta', linewidth=0.1);
plt.title("MOST REQUIRED SKILLS ")
plt.ylabel("SKILLS")
plt.xlabel("RATE");
```
